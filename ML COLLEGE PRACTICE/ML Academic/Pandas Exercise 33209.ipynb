## 5. Print each company's highest price car

```python
# Array for split datasets
companies = data.company.unique()
for i in range(len(companies)):
    print(companies[i], " =  ", data[companies[i]==data.company]["average-mileage"].max())
```

## 6. Find the average milage of each car making company

```python
# Array for split datasets
companies = data.company.unique()
for i in range(len(companies)):
    print(companies[i], " =  ", data[companies[i]==data.company]["average-mileage"].mean())
```
