## BAR CHART

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
name=['a','b','c','d']
mark=[67,58,92,78]
plt.bar(name,mark)
```

```python
name=['a','b','c','d']
mark=[67,58,92,78]
plt.bar(name,mark,label="Student Performance")
plt.xlabel("Student Name")
plt.ylabel("Stuent Marks")
plt.title("Student Report")
plt.legend()
```

```python
name=['a','b','c','d']
mark=[67,58,92,78]
plt.bar(name,mark,label="Student Performance",width=0.5,color='g')
plt.xlabel("Student Name")
plt.ylabel("Stuent Marks")
plt.title("Student Report")
plt.legend()
```

```python
name=['a','b','c','d','e']
f1=[77,66,85,86,72]
f2=[68,86,72,84,59]
plt.bar(name,f1,label="First Sem",width=0.5,color='g')
plt.bar(name,f2,label="Second Sem",width=0.5,color='r')
plt.xlabel("Student Name")
plt.ylabel("Stuent Marks")
plt.title("Student Report")
plt.legend()
```

```python
name=['a','b','c','d','e']
f1=[77,66,85,86,72]
f2=[68,86,72,84,59]
xps=np.arange(len(name))
plt.bar(xps,f1,label="First Sem",width=0.3,color='g')
plt.bar(xps+0.3,f2,label="Second Sem",width=0.3,color='r')
plt.xlabel("Student Name")
plt.ylabel("Stuent Marks")
plt.title("Student Report")
plt.legend()
```
