```python
#Convert all the elements of a numpy array from float to integer datatype

x=np.array([1.5,2.4,3,4.0,5])
print(x,"\n")
y = x.astype(int)
print(y)
```

```python
#Convert a binary numpy array (containing only 0s and 1s) to a boolean numpy array

x=np.array([0,1,1,0,1,0])
y=x%2!=0
print(x,"\n")
print(y)
```

```python
#Stack 2 numpy arrays horizontally i.e., 2 arrays having the same 1st dimension
#(number of rows in 2D arrays)

x = np.arange(1, 10).reshape(3, -1)
y = np.arange(10, 16).reshape(3, -1)
 
z = np.hstack((x, y))

print(x,"\n")
print(y,"\n")
print(z)
```

```python
#Stack 2 numpy arrays vertically i.e., 2 arrays having the same last dimension 
#(number of columns in 2D arrays)

x = np.arange(1, 10).reshape(-1, 3)
y = np.arange(10, 16).reshape(-1, 3)
 
z = np.vstack((x, y))

print(x,"\n")
print(y,"\n")
print(z)
```

Make use of NumPy to accept at least 20 numbers : Perform following
- Categorize as Odd and Even
- Print these categories using MATPLOTLIB
- Categorize as Prime and Non Prime
- Print these categories using MATPLOTLIB
- Identify Minimum / Maximum number 

```python
l=[int(i) for i in input().split()]
a = np.array(l)
print(a,"\n")

Even = a[a%2==0]
Odd =  a[a%2!=0]
print("Even Numbers : ",Even)
print("Odd Numbers : ",Odd)
print()

def is_prime(n):
    if n % 2 == 0 and n > 2: 
        return False
    return all(n % i for i in range(3, int(math.sqrt(n)) + 1, 2))

foo = np.vectorize(lambda x: is_prime(x))
primes = a[foo(a)] 

foo = np.vectorize(lambda x: not is_prime(x))
non_prime = a[foo(a)]
print("Prime Numbers : ",primes)
print("Non prime Numbers : ",non_prime)
print()

print("Maximum number : ", np.max(a))
print("Minimum number : ", np.min(a))

dict={"Prime":len(primes),"Non-prime":len(non_prime)}
number=list(dict.keys())
value=list(dict.values())

plt.subplot(1,2,1)
sns.barplot(x=number,y=value)
plt.title("Prime - Non-prime");

dict={"Even":len(Even),"Odd":len(Odd)}
number=list(dict.keys())
value=list(dict.values())

plt.subplot(1,2,2)
sns.barplot(x=number,y=value)
plt.title("Even - Odd");
```
