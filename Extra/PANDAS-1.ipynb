```python
import pandas as pd
import numpy as np
dict={
    "first":[12,13,np.nan,23],
    "second":[5,7,4,7],
    "third":[45,np.nan,57,np.nan]
}
df3=pd.DataFrame(dict)
df3
```

```python
df3.isnull().sum()
```

```python
column=['first', 'second', 'third']
for i in column:
    df3[i]=df3[i].fillna(df3[i].mean())
df3
```
