```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("company_sales_data.csv")
```

```python
# Plot 1

plt.plot(data["month_number"],data["total_profit"])
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total profit");
```

```python
# Plot 2

plt.plot(data["month_number"],data["total_profit"],color="r",linewidth=3,
         marker="o",linestyle="--",markerfacecolor='b',label="Profit data of last year")
plt.title("Company Sales data of last year")
plt.xlabel("Month number")
plt.ylabel("Sold units number")
plt.legend(loc="lower right");
```

```python
# Plot 3

plt.scatter(data["month_number"],data["toothpaste"],label="Tooth paste Sales data")
plt.title("Tooth paste Sales data")
plt.xlabel("Month Number")
plt.ylabel("Number of units sold")
plt.legend(loc="upper left")
plt.grid(linestyle="--")
```

```python
# Plot 4

plt.bar(data["month_number"]-.15,data["facecream"],width=0.3,label="Face Cream sales data")
plt.bar(data["month_number"]+.15,data["facewash"],width=0.3,label="Face Wash sales data")
plt.title("Facewash and facecream sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(loc="upper left")
plt.grid(linestyle="--")
```

```python
# Plot 5

plt.bar(data["month_number"],data["bathingsoap"])
plt.title("bathingsoap sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.grid(linestyle="--")
```

```python
# Plot 6

plt.hist(data["total_profit"],bins=6,rwidth=1,label="Profit data")
plt.title("Profit data")
plt.xlabel("profit range in dollar")
plt.ylabel("Actual Profit in dollar")
plt.legend(loc="upper left");
```

```python
# Plot 7

plt.stackplot(data["month_number"],data["facecream"],data["facewash"],data["toothpaste"],
             data["bathingsoap"],data["shampoo"],data["moisturizer"],
              labels=("Face Cream","Fash wash","Tooth paste","Bathing soap","Shampoo","Moisturizer")
             ,colors=["mediumvioletred","cyan","red","black","green","gold"])
plt.title("All product sales data using stack plot")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.legend(loc="upper left");
```
