```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.preprocessing import StandardScaler
```

```python
location="image.jpg"
```

```python
def Image_To_Numpy(location):
    img = imread(location)
    arr= np.zeros((28,28))
    image= cv2.resize(img,(28,28))
    for i in range(len(image)):
        for j in range(len(image)):
            blue=image[i,j,0]
            green=image[i,j,1]
            red=image[i,j,2]
            arr[i,j]=red*0.299+green*0.587+blue*0.114 
    arr= arr.reshape(-1)
    return arr
```

```python
gray_new=Image_To_Numpy(location)
```
